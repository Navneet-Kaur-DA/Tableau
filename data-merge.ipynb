```python
import pandas as pd
import numpy as np
from datetime import date
```

```python
dec13 = pd.read_csv("DATA/201312-citibike-tripdata_1.csv")
df1 = pd.DataFrame(dec13)
df1.head()
```

```python
dec18 = pd.read_csv("DATA/201812-citibike-tripdata.csv")
df2 = pd.DataFrame(dec18)
df2.head()
```

```python
df_all =  pd.concat([df1, df2], ignore_index=True)
df_all.head()
```

```python
df_all.tail()
```

```python
df_all.count()
```

```python
df_all['gender'].replace(to_replace=[0, 1, 2],
           value=['Unknown', 'Male', 'Female'],
           inplace=True)
```

```python
df_all.to_csv('DATA/2years_citibikenyc_tripdata.csv',  index=False, header=True)
```
